# 로또 정답 번호 별 빈도 파악

### 1. 자동 복권 번호 출력기 결과 20개 이상을 dataframe으로 만든 후 matplotlib으로 시각화하여 가장 많이 나온 번호 확인

##### 사용 라이브러리 : random, pandas, matplotlib

```python
import random


answer_list = []
count = int(input("로또를 몇 개 만들까요? :"))

for i in range(count):
    answer = []
    while len(answer) < 6:
        ans = random.randint(1,46)
        if ans in answer:
            pass
        else:
            answer.append(ans)
    answer_list.append(answer)
print(answer_list)
```

```python
import pandas as pd

df = pd.DataFrame(answer_list)
```

```python
df
```

```python
import matplotlib.pyplot as plt

all_values = df.values.flatten()

value_counts = pd.Series(all_values).value_counts()

plt.bar(value_counts.index, value_counts.values)
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.title('Histogram of lotto')
plt.show()
```
